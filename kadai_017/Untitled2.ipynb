```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >=20:
            print(f"{self.name}あなたは成人です")
        else:
            print(f"{self.name}あなたは未成年です")

human1 = Human("Taro",20)
human2 = Human("Jiro",14)

human_lists = [human1, human2]

for human_list in human_lists:
    human_list.check_adult()
```
